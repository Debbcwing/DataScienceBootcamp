## [Introduction](#P0)

Exceptions are special objects which are used to manage errors that arise during the execution of programs. Usually, exceptions will crash your program. However, Python gives you a convenient way to handle such errors through so-called exception handling. Exception handling helps you catch errors, and provide a way to deal with them so that your program doesn't crash. When you think that you have a code that could produce an error you should use exception handling.
Whenever we do something wrong, Python creates an exception object. If an unhandled exception occurs, your program will stop running and Python will show a stacktrace or traceback which includes the exact location where the error occurred and which error occurred.

Exceptions can be handled with try-except blocks:

```python
print(10 / 0)
```

```python
try:
    print(10 / 0)
except ZeroDivisionError:
    print("Can't divide by zero")
```

The try block tells Python to try a block of code. If that block of code runs successfully then the except block is skipped. If the try block is unsuccessful Python jumps to the except block, which tells Python what to do. When you use try-except blocks, your programs will continue running even if things start to go wrong. Instead of tracebacks, which can be confusing for users to read, you will see a friendly error message that you have written. If the error is not too serious, you can avoid crashing your program using try-except blocks.

## [Handling the Exceptions](#P0)

We know it's impossible to divide by zero, if we ask Python to do that it will raise an error.

```python
def normalize (a, b):
    sum = a+b
    return (a/sum, b/sum)

normalize(1,1) # (.5, .5)
normalize(0,0)
```

When this happens, Python stops the program and tells us what kind of exception has been raised. It also shows us where it happened in the form of a stack trace.

However, would it not be better to handle this error gracefully? We can use exception handling to modify our program, to tell Python what to do when this kind of exception occurs. That way, if it happens again, we’re prepared.

```python
def normalize (a, b):
    try:
        sum = a+b
        return (a/sum, b/sum)
    except ZeroDivisionError:
        return (0, 0)

normalize(1,1) # (.5, .5)
normalize(0,0) # (0, 0)
```

Ok, so far so good! But what if we want to do the following:

```python
normalize(1, "1")
```

It looks like adding a string and a number might not be such a good idea. That also means that except only catches the error you defined. You can handle this by using chained except clauses:

```python
def normalize (a, b):
    try:
        sum = a+b
        return (a/sum, b/sum)
    except ZeroDivisionError:
        return (0, 0)
    except TypeError:
        print("Wrong types")
        return (-1, -1)
    except:
        print("Another error has occurred")
```

```python
normalize(1,1)   # (.5, .5)
```

```python
normalize(0,0)   # (0, 0)
```

```python
normalize("a",0) # (0, 0)
```

The try-except statement works as follows:

- Attempt to execute the try clause
- If no exception occurs, we are done - skip all except clauses
- If exception occurs, go through all defined except handlers from top to bottom
    * If the exception's type matches, execute the except clause. If the exception handler does not raise another exception, we're done
    * if exception occurs but does not match the type, skip the except clause.

There is the special except without the name of the exception. This will catch any exception, and you should avoid it as much as possible.

## [Example - Preventing Crashes with Wrong User Inputs](#P0)

If we ask the users to give a number as an integer to our program and the user provide a letter instead we have to catch that error and let the user know that the program only accepts numbers:
For example:

```python
while 1:
    users_input = int(input("Write a number: "))
    print(users_input)
```

To fix this we have to add a try/except and and provide the ValueError as an exception:

```python
while 1:
    try:
        users_input = int(input("Write a number: "))
        print(users_input)
    except ValueError:
        print("Please use only numbers")
    else:
        break
```

Let's assume we have the simple app that divides two numbers and we have to handle ZeroDivisionError and show the result. That will look something like this:

```python
while 1:
    try:
        first_number = int(input("First number: "))
        second_number = int(input("Second number: "))
        answer = first_number / second_number
    except ValueError:
        print("Please use only numbers")
    except ZeroDivisionError:
        print("No zeros please")
    else:
        # we got the correct input - print answer, and exit the while 1 loop
        print(answer)
        break
```

Any code that depends on the try block succeeding is added to the else block. In this case if the division operation is successful, we use the else block to print the result.

### Example - FileNotFoundError exception

One common issue when working with files is handling missing files. The file you’re looking for might be in a different location, the filename may be misspelled, or the file may not exist at all. You can handle all of these situations in a straightforward way with a try-except block.

```python
filename = "constructor_students.txt"
file = open(filename)
# -> FileNotFoundError: [Errno 2] No such file or directory: 'constructor_students.exe'
```

To fix this we can use the FileNotFoundError exception.

```python
filename = "constructor_students.txt"
try:
    file = open(filename)
except FileNotFoundError:
    print("This file <{}> was not found".format(filename))
    file = open("defaults.txt", 'w')
```

Sometime we don't want to report every exception we catch. Sometimes we want our program to 'fail' silently when an exception occurs and continue on like nothing has happened. In this case we can use the pass statement.

```python
filename = "constructor_students.txt"
try:
    open(filename)
except FileNotFoundError:
    pass
```

Now when a FileNotFoundError is raised, the code in the except block runs, but nothing happens. No traceback is produced, and there’s no output in response to the error that was raised.

### Custom Exceptions

Let's create some custom exceptions by creating a light version of number guess game from the exercises we had.

```python
# define Python user-defined exceptions
class Error(Exception):
    """Base class for other exceptions"""
    pass

class ValueTooSmallError(Error):
    """Raised when the input value is too small"""
    pass

class ValueTooLargeError(Error):
    """Raised when the input value is too large"""
    pass

# our main program
# user guesses a number until he/she gets it right

# you need to guess this number
number = 10

while True:
    try:
        i_num = int(input("Enter a number: "))
        if i_num < number:
            raise ValueTooSmallError
        elif i_num > number:
            raise ValueTooLargeError
        break
    except ValueTooSmallError:
        print("This value is too small, try again!")
    except ValueTooLargeError:
        print("This value is too large, try again!")

print("Congratulations! You guessed it correctly.")
```

### Finally
The finally clause:

```python
try:
    raise ValueTooLargeError
except ValueTooLargeError:
    print("Uh uh! Value is too large!")
finally:
    print('Bye')
```

Finally always gets executed before leaving the try statement. Unhandled exceptions are re-raised after finally executes.
