## [Introduction](#P0)

So far we mainly focused on concepts that belong to the structured programming paradigm. Structured programming is a programming paradigm that makes heavy use of functions, block structures, and for loops, and as such improves the clarity, quality, and development time of your code.

This page aims at introducing a new concept called object-oriented programming, or OOP for short. OOP takes the ideas of structured development even further, by bundling functions/procedures and their required data together into a single unit called "object". Think of it like a real-world object, say a phone. A phone has certain data it stores such as phone numbers, alarms, or emails. At the same time, it offers different functionalities such as making a call, setting an alarm, or reading emails. Objects in OOP are very similar - they bundle certain functionalities (methods) and data (attributes) into one single object.

OOP is a very powerful paradigm, and will help you to structure your code much more logically. Therefore, it is one of the most common software development approaches today. Most programming languages you will encounter in your professional career have support for OOP.

## [The Class](#P0)

A class is a template or blueprint of an object. It describes how an object should behave and which data it should contain. Classes can be used to represent real-world things, but also more abstract things.

Let's have a look at how a blueprint for a car could look like:

```python
class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def start_engine(self):
        return "The " + self.brand + "'s engine just started"

    def stop_engine(self):
        return "The " + self.brand + "'s engine stopped"

    def makes_noise(self):
        return self.brand + "'s " + self.model + "makes a lot of noise"

    def makes_no_noise(self):
        return "A " + self.brand + " does not make any noise"
```

The class Car represents cars (duh!), but not just any. Rather, the class is structured in a way that it can represent different car models from different manufacturers, by using attributes like brand or model.

In addition to attributes, it also has some functionality such as start and stop the engine, make some noise (except maybe electric cars 😅). These functionalities are contained in so-called methods. Methods pretty much look like functions, but are tied to a certain class or object. They always contain the parameter ``self`` as the first parameter, but more on that in a few moments. Everything you learned so far about functions and attributes also applies to methods.

## [Objects - aka Creating an Instance of the Class](#P0)

Note, that so far we only created a blueprint for a car, and that we have not created an actual car. Let's have a look at how we can create an instance of a car.

Instances of classes are called objects. An instance of a car would look like this:

```python
porsche = Car('Porsche', '911')
```

```python
porsche.start_engine()
```

```python
porsche.model
```

```python
print(porsche.start_engine()) #  -> The Porsche's engine started
print(porsche.stop_engine())  #  -> The Porsche's engine stopped
```

By calling ``Car(...)`` with some arguments, we tell Python to create an actual instance of a car. The car instance has its own set of attributes, and you can call any of its methods just the way you would call a function. In our example, we tell Python to create a beautiful Porsche 911.

You probably noticed two things. First of all, to call a method you access it via ``object.method()``. Or in our case, via ``porsche.start_engine()``. Secondly, the method definition had an argument ``self``, which we do not use when calling the method on the object. When you call a function on an object, Python will automatically fill the ``self`` argument, and place the object itself in there.

To give you an illustration of what happens, look at the following example. Both calls are exactly the same:

```python
# The way to write method calls
porsche.start_engine()
```

```python
# Which will be "translated" by Python like the following
Car.start_engine(porsche)
```

## [The ``__init__`` method](#P0)

Python classes have a few special methods that have a special purpose. The most important is the ``__init__`` method.
The ``__init__(self)`` is the constructor of the class. The constructor will be run by Python every time we create an object (creating an object is the same as creating an instance of a class). ``__init__`` is always the first method called on an object, and is used to initialize the object. Most notably, you define all attributes of the class in the constructor, and assign initial values to them.
As with all methods, the parameter ``self`` is always required. Python will automatically place the newly created object before calling the constructor. In our case, we tell Python to create a new car by calling ``Car('Porsche', '911')``. We are calling the Car class like a function. Python will then run the ``__init__()`` function of Car, with a reference to the newly created object (i.e self = porsche) as its argument. You can pass additional parameters after the ``self``. In the example of the Car class, we pass a brand and a model as additional parameters.
Within a method or constructor, ``self`` gives you access to the attributes and other methods of this object. Recall that an object is a bundle of data and functions. That also means, you can write new values ``self.brand = brand`` in one method, and then later access the same date with ``self.brand`` in any other method.
Note, the ``__init__()`` method has no explicit return statement, but Python automatically returns an instance.

## [Class Attribute References](#P0)

So far you have seen instance attributes, that have been declared with the self accessor in the constructor, e.g. ``self.brand = brand``. Instance attributes can only be accessed by instance objects, not by the class itself:

```python
# instance attributes can only be queried on instances
audi = Car('Audi', 'RS6')
audi.brand # -> Audi
```

```python
# but not on the class itself
Car.brand # error
```

There is also the possibility to define attributes on the class instead of on the instance. These are called class attributes. For example:

```python
class MyClass:
    class_attr = [1, 2] # this is the class attribute
    def __init__(self, my_attr):
        self.my_attr = my_attr # this is the instance attribute
```

```python
my_instance = MyClass('hello')
```

```python
print(my_instance.my_attr)  # hello
```

```python
print(my_instance.class_attr)  # [1,2]
```

```python
print(MyClass.class_attr) # [1,2]
```

Let's analyze some different applications for class attributes.

**1. Storing constants:** Class attributes can be accessed as attributes of the class itself.

```python
class Circle:
    PI = 3.1415

    def __init__(self, rad):
        self.rad = rad

    def area(self):
        return Circle.PI * self.rad * self.rad

print(Circle.PI)  # 3.1415
new_circle = Circle(10)
print(new_circle.PI)  # 3.1415
print(new_circle.area())  # 314.150
```

**2. Defining default values:** In this case we can set an upper limit on the amount of data we want to handle inside a list or a database. Let's take a look at an example:

```python
class LimitedList:
    limit = 15

    def __init__(self):
        self.data = []

    def add(self, item):
        if len(self.data) > self.limit:
            raise Exception("We can not save any more data")
        self.data.append(item)


print(LimitedList.limit) # 15
```

For every instance of ``LimitedList`` we create, we can change the limit like this:

```python
my_limit = LimitedList()
other_limit = LimitedList()
```

```python
my_limit.limit = 20
```

```python
print(my_limit.limit) # -> 20
print(other_limit.limit)  # -> 15
```

``my_limit`` can now handle maximum 20 items and the other instances can handle 15.

This only makes sense if you want your typical instance of limit to hold just 15 elements or fewer. If you want to give each instance a different limit then it does not make sense to use a class attribute. You should define it as an instance variable. If changing the limit is just an exception then you should use class attributes.

**3. Tracking all data across all instances of the given class:**

Let's create a class called ``Students`` and record everyone's names. ``Name`` will be a class attribute.

```python
class Students:
    names = []

    def __init__(self, name):
        self.name = name
        Students.names.append(name)
```

```python
ilya = Students('Ilya')
```

```python
sophia = Students('Sophia')
```

```python
print(Students.names) # ['Ilya', 'Sophia']
```

```python
print(sophia.names) # ['Ilya', 'Sophia']
```

## [Class and Instances Attributes](#P0)

**IMPORTANT NOTE**

In the example above, we use ``Students.names`` instead of ``self.names``. Why this? Look at this example:

```python
class Teachers:
    count = 0

    def __init__(self, name):
        self.name = name
        self.count = self.count + 1

simon = Teachers('Simon')
manuel = Teachers('Manuel')
print(Teachers.count) # -> 0
print(simon.count) # -> 1
```

Ooops! Shouldn't count be 2? What happened here?

Using ``self.count`` to read will give you access to the class variable.

Writing ``self.count`` will create a new instance variable, that will shadow the class variable.

This is a very common pitfall - make sure to use ``Teachers.count = Teachers.count + 1`` instead if you want to use shared information across instances.

Let's look at a couple of related cases:

```python
class Dog:
    kind = 'German Shepherd' # class variable shared by all instances

    def __init__(self, name):
        self.name = name    # instance variable unique to each instance


g = Dog('Spike')
print(g.kind) # German Shepherd
print(g.name) # Spike
```

**WARNING!**

```python
class Dog:
    kind = 'German Shepherd'  # class variable shared by all instances
    trick = []

    def __init__(self, name):
        self.name = name  # instance variable unique to each instance

    def add_tricks(self, trick):
        self.trick.append(trick)


dog1 = Dog('Spike')
dog1.add_tricks('roll over')

dog2 = Dog('Lassie')
dog2.add_tricks('play dead')

print(dog1.name) # Spike
print(dog2.name) # Lassy

print(dog1.trick) # ['roll over', 'play dead']
```

The dogs share the same tricks even if some of them are not theirs.

To fix this we have to change the way we constructed the class.

```python
class Dog:
    kind = 'German Shepherd'  # class variable shared by all instances

    def __init__(self, name):
        self.name = name  # instance variable unique to each instance
        self.trick = []

    def add_tricks(self, trick):
        self.trick.append(trick)


dog1 = Dog('Spike')
dog1.add_tricks('roll over')

dog2 = Dog('Lassie')
dog2.add_tricks('play dead')

print(dog1.name) # Spike
print(dog2.name) # Lassy

print(dog1.trick) # ['roll over']
```

**Additional remarks about instantiation and objects**

- In general, anything in Python is an object. Even a class is an object. When you define a new class (e.g. class Dog), Python creates a new class object which supports instantiation (``Dog('Spike')``), and attribute reference (``Dog.kind = 'German Shepherd'``).
- When you instantiate a class object ``dog1 = Dog('Spike')``, Python creates a new instance object based on that class. Instance objects only support attribute reference (e.g. dog1.kind), but not instantiation (e.g. ``dog1("Paul")`` does not work).
- If you do not yet fully understand what it means, do not worry. You will as you get more experienced.

## [Inheritance](#P0)

Inheritance is a fundamental concept of object-oriented programming. It enables new classes or objects to receive - aka inherit - properties and methods from existing classes.

To briefly recap:

     - an object is a self-contained collection of properties and methods that belong together logically.
     - a class is a template to build instances of objects.
     - every object is built from a class.

Inheritance is a way to express the relationship between classes. It is a way to instruct Python that you want to build a new class that is similar to a pre-existing class, but with some modifications. You can do that by creating a new class from scratch (and replicating a lot of code), but a more efficient way to implement that is to reference to an existing class, copy all of its methods over, and simply indicate what is different.

There are two main concepts related to class inheritance:

- **Subclassing** lets you make a new class based on a previous one
- **Overriding** lets you change certain parts of that class.

Inheritance also organizes your objects into a hierarchy, which often makes code easier to understand. Most importantly, inheritance allows you to reuse, structure and organize your code more effectively.

### Subclassing

A class can inherit from any other class with one simple statement. In the following example, the subclass ChildClass inherits all the methods and properties from its superclass ParentClass:

```python
class ChildClass(ParentClass):
    pass
```

Classes that are used as bases for inheritance are known as superclasses, base classes, or parent classes. A class that inherits from a superclass is called a subclass, derived class, or child class. A child always inherits visible properties and methods from its parent, while adding additional properties and methods of its own.

Subclassing is not limited to creating one level: a child class can be the parent of another class. In this case, in each level we end up with a more specific version of a class. For example - a shepherd is a dog, a dog is a mammal, and a mammal is an animal. A lemon is a citrus fruit, which is a fruit. In such cases inheritance makes a lot of sense.

If there is no relationship between a subclass and a superclass, you are probably doing something wrong and you should not use inheritance. For example, a car has a door, but it does not make sense to say that the door is a special kind of car or the other way around. These two items form a composition relation rather than an inheritance relation.

Let's take another look at our car example. Say we would like to create an electric car class:

```python
class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def start_engine(self):
        return "The " + self.brand + "'s engine just started"

    def stop_engine(self):
        return "The " + self.brand + "'s engine stopped"

    def make_noise(self):
        return self.brand + "'s " + self.model + " makes a lot of noise"
```

```python
class ElectroCar(Car):
    def make_noise(self):
        return "A " + self.brand + " is an electric car and does not make any noise"

    def recharge(self):
        return "Recharging..."
```

```python
mustang = Car('Ford', 'Mustang')
mustang.start_engine() # The Ford's engine just started
mustang.make_noise()   # Ford's Mustang makes a lot of noise
```

```python
tesla = ElectroCar('Tesla', 'Roadster')
tesla.start_engine() # The Tesla's engine just started
tesla.make_noise()   # A Tesla is an electric car and does not make any noise
```

What happened here?

- ``ElectroCar`` is a subclass of ``Car``.
- When declaring ``ElectroCar``, Python creates a new class, takes all the methods from the superclass ``Car``, and assigns them to the new class. The methods ``__init__``, ``start_engine()``, and ``stop_engine`` can be used in ``ElectroCars`` right away - no need to declare them again.
- If you need different implementations of certain methods, just override them. Because electric cars make less noise, we have overridden ``make_noise()`` in our example.
- And of course, you can add new methods like ``recharge()`` in our example.

### Specialising the Child Class

Sometimes you need to specialize certain methods to differentiate the child class from its parent. For example, we can add new attributes to our ElectroCar.
Electric cars run on batteries instead of gasoline, so we should add a battery as an attribute and define some methods for it. We will store the battery size along with a method that prints out the battery size:

```python
class ElectroCar(Car):
    def __init__(self, brand, model):
        super().__init__(brand, model)
        self.battery = 60

    def battery_size(self):
        return f'This car has {self.battery} kWh battery'


tesla = ElectroCar('Tesla', 'Roadster')
print(tesla.battery_size()) # This car has 60 kWh battery
```

We added a ``battery`` attribute to the ``ElectroCar`` class, which means that this attribute will be associated with all the instances created from the class ``ElectroCar``, but not for instances created from the class ``Car``. The same thing will happen with the method ``battery_size()``.
Also, we have overridden the ``__init__`` with our new code, because we wanted to add a battery to the Electro car class. In this particular case we still wanted the attributes from the Car class but we needed an additional attribute. Instead of replicating the code from ``Car.__init__``, we can use the ``super()`` function. This function gives you access to the superclass.
Of course, you don't need to use ``super()`` if it does not make sense to you. For example if the ``Car`` class has a method ``fill_tank()`` the ``ElectroCar`` class does not really need that, so we can override it saying The electric car does not need fuel.

### Modularisation

The more your code grows, the more classes with different attributes and methods you will have. One way to break up this verbosity is to see if some parts of the class can be grouped together and written as separate classes. This allows you to break a class into smaller classes that work together.

If we add more details to the ``ElectroCar`` class we may find ourselves adding too many attributes and methods only for the battery of the car. It would make more sense to create a new class specifically for the battery:

```python
class Battery:
    def __init__(self):
        self.battery = 60

    def battery_size(self):
        return f'This car has {self.battery} kWh battery'


class ElectroCar(Car):
    def __init__(self, brand, model):
        super().__init__(brand, model)
        self.battery = Battery()


tesla = ElectroCar('Tesla', 'Roadster')
print(tesla.stop_engine())
print(tesla.battery.battery_size())
```

Let's add another method to the ``Battery`` class that describes the range of the car based on its battery size:

```python
class Battery:
    def __init__(self):
        self.battery = 60
        self.range = None

    def battery_size(self):
        return f'This car has {self.battery} kWh battery'

    def get_range(self):
        if self.battery == 60:
            self.range = 240
        elif self.battery == 85:
            self.range = 270
        return f'This car can go {str(self.range)} km in full charge.'


class ElectroCar(Car):
    def __init__(self, brand, model):
        super().__init__(brand, model)
        self.battery = Battery()


tesla = ElectroCar('Tesla', 'Roadster')
print(tesla.battery.get_range()) # This car can go 240 km in full charge.
```
