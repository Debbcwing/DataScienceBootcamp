```python
import numpy as np
```

```python
sf = [220, 183, 301, 199, 323] 
bst = [125, 186, 156, 133, 202]
```

```python
sf_mean = np.mean(sf)
bst_mean = np.mean(bst)
n1 = len(sf)
n2 = len(bst)
```

```python
sf_s = np.var(sf, ddof=1)
bst_s = np.var(bst, ddof=1)
# the normal formula for variance is 1/n * sum(x-x_mean)^2, but here we use 1/(n-1) * sum(x-x_mean)^2 as an estimator
# the mathematical reason for that is a bit more complicated but the summary is that we dont get an unbiased estimator by dividing by n
# we get numpy to use this formula by including the argument ddof = 1 
```

```python
s_p = np.sqrt(((n1-1)*sf_s+(n2-1)*bst_s)/(n1+n2-2))
```

```python
t = (sf_mean - bst_mean) / (s_p*np.sqrt(1/n1 + 1/n2))
```

```python
print(f"The t-value obtained is {t=}")
```
